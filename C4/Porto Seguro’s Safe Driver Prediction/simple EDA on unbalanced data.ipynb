```python
# https://www.cnblogs.com/cgmcoding/p/13355748.html   sns.countplot 应用类别特征， 只传入xy其中一个  
# https://www.cnblogs.com/cgmcoding/p/13384535.html   sns.distplot 应用数值特征(连续) 直方图
# https://www.cnblogs.com/cgmcoding/p/13356745.html   sns.barplot xy全要传入

def bar_plot(col, data, hue=None):
    f, ax = plt.subplots(figsize=(12,5))
    sns.countplot(x=col, hue=hue, data=data)
    plt.show()

def dist_plot(col, data):
    f, ax = plt.subplots(figsize=(12,5))
    sns.distplot(data[col].dropna(),kde=False, bins=10)
    plt.show()
    
def bar_plot_ci(col, data):
    f, ax = plt.subplots(figsize=(12,5))
    sns.barplot(col, "target", data=data)
    plt.show()
```

```python
# 不平衡的数据不适合用accuracy，因为accuracy一直都是很高的
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
labels = train.columns[2:]

X = train[labels]
y = train["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

model = XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
```

```python
print("accuracy: %.2f" % (accuracy * 100))
```

```python
model = XGBClassifier()
model.fit(X_train[["ps_calc_01"]], y_train)
y_pred = model.predict(X_test[["ps_calc_01"]])

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f" % (accuracy * 100))
```

```python
# 没区别
```

```python
from sklearn.metrics import confusion_matrix

conf_mat = confusion_matrix(y_true=y_test, y_pred=y_pred)
print(conf_mat)
```

```python
labels = ["Class 0", "Class 1"]
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(conf_mat, cmap=plt.cm.Blues)
fig.colorbar(cax)
```

```python
# 拟合效果不是很好，可以对数据重新采样
```

```python
count_class_0, count_class_1 = train.target.value_counts()
```

```python
df_class_0 = train[train["target"] == 0]
df_class_1 = train[train["target"] == 1]
```

```python
print(count_class_1, count_class_0)
```

### random under-sampling

```python
df_class_0_under = df_class_0.sample(count_class_1) # 下采样
df_class_0_under.shape
```

```python
df_class_0_under.head()
```

```python
# 平衡数据组合一起
df_test_under = pd.concat([df_class_0_under,df_class_1], axis=0)
print(df_test_under.shape)
print(df_test_under.target.value_counts())
```

```python
df_test_under.target.value_counts().plot(kind="bar")
```

### random over-sampling

```python
df_class_1_over = df_class_1.sample(count_class_0, replace=True) # 上采样
df_class_1_over.shape
```

```python
df_test_over = pd.concat([df_class_1_over, df_class_0], axis=0)
print(df_test_over.target.value_counts())
```

```python
df_test_over.target.value_counts().plot(kind="bar")
```
