```python
from collections import Counter

# 数据集大，分批读取数据
size = 10000
counter = Counter()
groupbydata = pd.DataFrame()
for chunk in pd.read_csv("./kesai/input/test_set.csv", chunksize=size):
    tmp = chunk.groupby("age").size().reset_index().rename(columns={0:"age_count"}) # 增加age_count特征
    groupbydata = pd.concat([groupbydata, tmp])
print(groupbydata.head())
print(groupbydata.shape)
```
